```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

```python
df = pd.read_csv(r"C:\Retail Sales Performance Analysis\retail_sales_dataset.csv")
df.head()
```

```python
df.shape
df.columns
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
category_sales = df.groupby("category")["sales_amount"].sum()
plt.figure(figsize=(8,5))
category_sales.plot(kind="bar", color="skyblue")
plt.title("Category-wise Sales")
plt.xlabel("Category")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
brand_sales = df.groupby("brand")["sales_amount"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
brand_sales.plot(kind="bar", color="orange")
plt.title("Brand-wise Sales")
plt.xlabel("Brand")
plt.ylabel("Sales Amount")
plt.xticks(rotation=90)
plt.show()
```

```python
region_sales = df.groupby("region")["sales_amount"].sum()
plt.figure(figsize=(5,4))
plt.pie(region_sales,
        labels=region_sales.index,
        autopct="%1.1f%%",
        startangle=90)
plt.title("Region-wise Sales")
plt.show()
```

```python
sns.countplot(data=df, x="customer_gender")
plt.title("Customer Gender Distribution")
plt.show()
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="customer_segment")
plt.title("Customer Segment")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(data=df, x="payment_method")
plt.title("Payment Method")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(3,2))
sns.countplot(data=df, x="sales_channel")
plt.title("Sales Channel")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(df["sales_amount"], bins=20)
plt.title("Sales Amount Distribution")
plt.show()
```

```python
top_products = df.groupby("product_name")["sales_amount"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(5,3))
top_products.plot(kind="bar", color="green")
plt.title("Top 10 Products")
plt.xlabel("Product Name")
plt.ylabel("Sales Amount")
plt.xticks(rotation=90)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)
plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df["transaction_date"] = pd.to_datetime(df["transaction_date"])
df["Month"] = df["transaction_date"].dt.to_period("M").astype(str)
monthly_sales = df.groupby("Month")["sales_amount"].sum()
plt.figure(figsize=(8,3))
monthly_sales.plot(marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
top_brands = df.groupby("brand")["sales_amount"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(6,4))
top_brands.plot(kind="bar", color="red")
plt.title("Top 10 Brands")
plt.xlabel("Brand")
plt.ylabel("Sales Amount")
plt.xticks(rotation=90)
plt.show()
```
